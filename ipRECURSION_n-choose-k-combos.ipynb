```python
def find_combinations(n, k, slate=[], results=[]):
    if n == 0:
        if len(slate) == k:
            results.append(sorted(slate.copy()))
        return
    find_combinations(n-1, k, slate, results)
    slate.append(n)
    find_combinations(n-1, k, slate, results)
    if slate:
        slate.pop()
    return results

n = 15
k = 5
sorted(find_combinations(n, k))

# n = 6
# k = 6
# sorted(find_combinations(n, k))
```
